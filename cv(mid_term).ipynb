```python
!pip install opencv-python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
# READ THE IMAGE
image = cv2.imread('image.png')
```

```python
#DISPLAY ORIGINAL IMAGE
plt.imshow(image)
```

```python
# TRANSLATION ON RIGHT SIDE
def translate_right(image, shift):
    rows, cols = image.shape[:2]
    M = np.float32([[1, 0, shift], [0, 1, 0]])
    translated_image = cv2.warpAffine(image, M, (cols, rows))
    return translated_image

translated_right = translate_right(image, 100)  # Shift by 100 pixels to the right
```

```python
#TRANSLATED RIGHT SIDE
plt.imshow( translated_right)
```

```python
# RESIZE THE IMAGE TO DOUBLE OF THE ORIGINAL
resized_double = cv2.resize(image, None, fx=2, fy=2, interpolation=cv2.INTER_LINEAR)
```

```python
# RESIZE THE IMAGE TO DOUBLE OF THE ORIGINAL
plt.imshow( resized_double)
```

```python
# SHEAR ON Y_AXIS
def shear_y(image, shear_factor):
    rows, cols = image.shape[:2]
    M = np.float32([[1, shear_factor, 0], [0, 1, 0]])
    sheared_image = cv2.warpAffine(image, M, (cols, rows))
    return sheared_image

sheared_y = shear_y(image, 0.5)  # Shear by 0.5 on y-axis
```

```python
# SHEAR ON Y_AXIS
plt.imshow(sheared_y)
```

```python
# TRANSLATION FUNCTION
def my_translation(image, tx, ty):
    rows, cols = image.shape[:2]
    M = np.float32([[1, 0, tx], [0, 1, ty]])
    translated_image = cv2.warpAffine(image, M, (cols, rows))
    return translated_image

translated_custom = my_translation(image, 50, 50)
```

```python
# TRANSLATION FUNCTION
plt.imshow( translated_custom)
```
